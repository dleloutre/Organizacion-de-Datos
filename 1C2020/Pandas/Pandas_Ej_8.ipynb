```python
#A. Indicar los títulos de los productos de la categoría "Men" para los cuales el Importe de
#venta supera el promedio mensual de ventas de los productos de la misma categoría.
#(por ejemplo si el promedio de Abril de "Men" es 120 dolares y un producto se vendió en
#Abril a 135 dolares lo tenemos que listar). Usar Transform.
```

```python
productos=pd.read_csv('products.csv')
productos
```

```python
sales = pd.read_csv('sales.csv')
sales
```

```python
productos_y_ventas = productos.merge(sales,on='productId',how='inner')
productos_y_ventas
```

```python
productos_y_ventas['promedio_x_categoria'] = productos_y_ventas.groupby('category')['importe(usd)'].transform('mean')
productos_y_ventas
```

```python
productos_y_ventas[(productos_y_ventas['category'] == 'Men') & (productos_y_ventas['importe(usd)'] > productos_y_ventas['promedio_x_categoria'])].title.unique()
```

```python
#B. Indicar el top-10 de productos que se vendieron mayor cantidad de días de forma
#consecutiva.
```

```python
#productos_y_ventas.drop(columns=['category','description','promedio','importe(usd)','promedio_x_categoria'],inplace=True)
```

```python
productos_y_ventas=productos_y_ventas.rename(columns={'anio':'year','mes':'month','dia':'day'})
productos_y_ventas['fecha']=pd.to_datetime(productos_y_ventas[['year','month','day']])
productos_y_ventas.drop(columns=['day','month','year'],inplace=True)
```

```python
productos_y_ventas.drop(columns=['category','description','importe(usd)','promedio_x_categoria'],inplace=True)
```

```python
productos_y_ventas
```

```python
productos_y_ventas = productos_y_ventas.sort_values(by='fecha', ascending=True)
productos_y_ventas
```

```python
#(lo saqué de la internet) primero agrupo por productId calculando la diferencia entre fechas. La diferencia
#vuelve en formato datetime asi que me quedo con los dias.. los que no tienen diferencia porque hay un unico valor
#le asigno un 1 (un solo dia consecutivo). ne(1) devuelve True si el valor de la serie es distinto de uno (si no false)
#a los q no son iguales a 1 (o sea que no son consecutivos) se les calcula la suma acumulativa
#entonces dentro de cada productId, los que son dias consecutivos van a tener el mismo valor "s", los demás no (van
#a tener valores crecientes). Lo de recién sería como una columna nueva, ahora si agrupo por productId y por la 
#nueva columna, haciendo size() obtengo la cantidad de dias consecutivos para cada productId. Size calcula el 
#tamaño de cada grupo
s = productos_y_ventas.groupby('productId').fecha.diff().dt.days.fillna(1).ne(1).cumsum()
productos_y_ventas.groupby(['productId', s]).size().reset_index(level=1, drop=True).sort_values(ascending=False).head(10)
```
