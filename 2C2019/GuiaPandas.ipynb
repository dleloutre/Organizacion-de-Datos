Dado el registro de notas de los alumnos de la forma (padrón, materia, nota, fecha). Se pide
resolver utilizando Pandas:
A. Cuál es el promedio de notas
B. Cuál es la nota más alta y la nota más baja registrada durante el año 2019.
C. Cuál es el padrón con mayor cantidad de materias aprobadas durante el último
cuatrimestre.
D. Cuál es la nota promedio
E. Cuál es la nota promedio por padrón.

```python
import pandas as pd
```

```python
df = pd.read_csv('notas.csv')
df
```

```python
#promedio de notas
df['nota'].mean()
```

```python
#nota más alta y la nota más baja registrada durante el año 2019
#primero convierto a fecha y saco el año

df['fecha_nota'] = pd.to_datetime(df['fecha'])
df['año'] = df['fecha_nota'].dt.year
df
```

```python
#nota mas alta
notas_2019 = df[df['año']==2019]
notas_2019.max()
```

```python
notas_2019.min()
```

```python
df['mes'] = df['fecha_nota'].dt.month
df
```

```python
#padrón con mayor cantidad de materias aprobadas durante el último cuatrimestre
notas_2019['mes'] = notas_2019['fecha_nota'].dt.month
notas_2019
```

```python
notas_ult_cuatr = notas_2019[notas_2019['mes']>4]
notas_ult_cuatr
```

```python
#tendria que agrupar para cada padron la cantidad de notas mayores a 4
materias_aprobadas = notas_ult_cuatr[notas_ult_cuatr['nota']>=4]
alumnos_aprobados= materias_aprobadas.groupby(['padron'],as_index=False).agg({'materia':'count'})
alumnos_aprobados.columns = ['padron', 'cantidad_materias_aprobadas']
alumnos_aprobados
```

```python
padron_mas_materias_aprobadas = alumnos_aprobados.sort_values(by= 'cantidad_materias_aprobadas', ascending=False)
padron_mas_materias_aprobadas.iloc[0]
```

```python
#cual es la nota promedio por padron
notas_promedio_por_padron = df.groupby(['padron'],as_index=False).agg({'nota':'mean'})
notas_promedio_por_padron.columns = ['padron', 'nota_promedio']
notas_promedio_por_padron
```
